```python
import pandas as pd
import re
from rank_bm25 import *
from nltk.tokenize import word_tokenize
from gensim.parsing.preprocessing import STOPWORDS
from nltk.stem import WordNetLemmatizer
import nltk
# nltk.download('omw-1.4')
# nltk.download('wordnet')
# nltk.download('punkt')
```

```python
def special_chars_removal(lst):
    lst1=list()
    for element in lst:
        str=''
        str = re.sub(r'[^\w\s]','',element)
        lst1.append(str)
    return lst1
```

```python
def stopwords_removal_gensim_and_lemmatize(lst):
    lst1=list()
    for str in lst:
        text_tokens = word_tokenize(str)
        tokens_without_sw = [lemmatize(lemmatizer, word) for word in text_tokens if not word in STOPWORDS]
#         str_t = ' '.join(tokens_without_sw)
        lst1.append(tokens_without_sw)
 
    return lst1
```

```python
lemmatizer = WordNetLemmatizer()
def lemmatize(lemmatizer, word):
    return lemmatizer.lemmatize(word)
```
